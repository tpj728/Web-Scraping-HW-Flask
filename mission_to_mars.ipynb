```python
soup = bs(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('div', class_="slide")
```

```python
results[1].text
```

```python
titles = []
paragraphs = [] 
for result in results:
    try:
        title = result.find('div', class_="content_title").text 
        paragraph = result.find('div', class_="rollover_description_inner").text
        titles.append(title)
        paragraphs.append(paragraph)
        
        if (title and paragraph):
            print(title)
            print(paragraph)
    except AttributeError as e:
        print(e)
```

```python
def init_browser():
    executable_path={"executable_path":"/usr/local/bin/chromedriver"}
    return Browser("chrome", **executable_path, headless=False)
```

```python
def scrape():
    browser = init_browser()
    
    url1 = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
    browser.visit(url1)
    
    html = browser.html
    soup1 = bs(html, "html.parser")
    
    image_url = soup1.find("article", style_="background_image").text()

    return image_url
```

```python
scrape()
```

```python
url4 = 'https://space-facts.com/mars/'
response4 = requests.get(url4)
soup4 = bs(response4.text,'html.parser')
```

```python
print(soup4.prettify())
```

```python
results4 = soup4.find_all('td', class_="column-1")
```

```python
results5 = soup4.find_all('td', class_="column-2")
```

```python
results4[0]
```

```python
headers = []
facts = []
for result in results4:
        
    try:
        header = result.text 
        headers.append(header)
        
        if header:
            print(header)

    except AttributeError as e:
         print(e)

for result in results5:
        
    try:
        fact = result.text 
        facts.append(fact)
        
        if fact:
            print(fact)

    except AttributeError as e:
         print(e)
```

```python
import pandas as pd
```

```python
d={'Planet Profile':headers,'Fact': facts}
Mars_facts = pd.DataFrame(d,columns=['Planet Profile','Fact'])
```
